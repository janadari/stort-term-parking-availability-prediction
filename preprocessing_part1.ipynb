```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Import data
df = pd.read_csv('D:/research/mydataset/myset.csv')
df.head(10)
```

```python
df['concat_date_time_in']= df['date']+" "+df['in']
df['concat_date_time_in']= pd.to_datetime(df['concat_date_time_in'],format = "%m/%d/%Y %H:%M")
df['concat_date_time_out']= df['date']+" "+df['out']
df['concat_date_time_out']= pd.to_datetime(df['concat_date_time_out'],format = "%m/%d/%Y %H:%M")
df['time_diff'] = (df['concat_date_time_out']-df['concat_date_time_in']).astype('timedelta64[m]')
df.head(10)
```

```python
df.Type[df.Type =='car'] =1
df.Type[df.Type =='van'] =2
df.Type[df.Type =='van '] =2
df.Type[df.Type =='jeep'] =3
df.Type[df.Type =='bike0'] =4
df.Type[df.Type =='threeweel'] =5
df.Type[df.Type =='bike'] =6
df.Type[df.Type =='bike '] =6
df.Type[df.Type =='lorry'] =7
df.Type[df.Type =='cab '] =8
df.Type[df.Type == 'cab'] = 8

df.codeno[df.codeno =='A'] =1
df.codeno[df.codeno =='B'] =2
df.codeno[df.codeno =='c'] =3


df.head(10)
```

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('D:/research/mydataset/7.csv')
df.head(10)
```

```python
df.describe()
```

```python
df = df[['codeno','location','concat_date_time_in','concat_date_time_out','time_diff','day']]
df.head(10)
```

```python
df ['Dates'] = pd.to_datetime(df['concat_date_time_in']).dt.date
df['Time'] = pd.to_datetime(df['concat_date_time_in']).dt.time
df.head(10)
```

```python
import matplotlib.dates as mdates
plt.plot([],[])
park_name = df['codeno'].unique()
#print(park_name)
for i in range(len(park_name)):
    s = park_name[i]
    rate = df[df['codeno'] == s]['time_diff']
#     time = df[df['codeno'] == s]['Time']
    time=pd.to_datetime(df[df['codeno'] == s]['Time'],format='%H:%M:%S')
    plt.scatter(time,rate,label=s)
plt.gcf().autofmt_xdate()
myFmt = mdates.DateFormatter('%H:%M')
plt.gca().xaxis.set_major_formatter(myFmt)    
plt.xlabel('Time')
plt.ylabel('Time Different')
plt.show()
plt.close()
```

```python
df.head(10)
# count_in = df.groupby('concat_date_time_in','time_diff')['concat_date_time_in','time_diff'].count()
count_in = df.groupby(['concat_date_time_in','location','day'], as_index=True)['concat_date_time_in'].count()
            

count_out = df.groupby(['concat_date_time_out','location','day'], as_index=True)['concat_date_time_out'].count()


count_data = pd.concat([count_in, count_out], axis=1).fillna(0).cumsum()
dt = count_data['concat_date_time_in'] - count_data['concat_date_time_out']
print(dt)
```

```python
df = dt.to_frame().reset_index()
df
df.columns = ['time', 'location' ,'day','occupancy']
df
```

```python
df['Dates'] = pd.to_datetime(df['time']).dt.date
df['Time'] = pd.to_datetime(df['time']).dt.time
df
```

```python
import matplotlib.dates as mdates
plt.plot([],[])
park_name = df['location'].unique()
#print(park_name)
for i in range(len(park_name)):
    s = park_name[i]
    rate = df[df['location'] == s]['occupancy']
#     time = df[df['codeno'] == s]['Time']
    time=pd.to_datetime(df[df['location'] == s]['Time'],format='%H:%M:%S')
    plt.scatter(time,rate,label=s)
plt.gcf().autofmt_xdate()
myFmt = mdates.DateFormatter('%H:%M')
plt.gca().xaxis.set_major_formatter(myFmt)    
plt.xlabel('Time')
plt.ylabel('Occupancy Rate')
plt.show()
plt.close()
```
